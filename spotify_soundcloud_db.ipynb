# The Cleaning and Manipulation of The Spotify Top 200 Chart csv
In the cells below we:
  
    - Filtered the data, keeping the chart information from only the years 2019-2021
    - Narrowed down the region, only showing the top charts from The United States
    - Got rid of any unneccessary information such as the "Viral50" information 
    - Narrowed it down to the top 50 songs at the first of every month from 2019-2021 

```python
#view your data
spotify_charts
```

```python
#narrow down date, only showing from January of 2019 to present
spotify_charts = spotify_charts.drop(labels=range(0, 5723700), axis=0)
```

```python
#limit the region to the top songs in The United States
usa_chart= spotify_charts.loc[spotify_charts['region']=='United States']
USA = pd.DataFrame(usa_chart)
```

```python
#get rid of "Viral50" lists because it is not relevant to our charts 
only_top_200= USA[USA["chart"].str.contains("viral50")==False]
```

```python
only_top_200=pd.DataFrame(only_top_200)
```

```python
#get rid of trend column because it does not add any useful data to our analysis 
del only_top_200['trend']
```

```python
#drop any missing rows NAN
only_top_200.dropna()
```

```python
#to better read our data, we are limiting it to the top 50 instead of the top 200
top_50_only = only_top_200.drop(only_top_200["rank"].loc[only_top_200["rank"] > 50].index)
```

```python
top_50_only= pd.DataFrame(top_50_only)
```

```python
#only keep the 1st of every month 
#this will allow us to see the top 50 of each month of the year
#still provides us with 1700 rows of data 
monthly= top_50_only.loc[(top_50_only['date'] == '2019-01-01') | 
                         (top_50_only['date'] == '2019-02-01')| 
                         (top_50_only['date'] == '2019-03-01')| 
                         (top_50_only['date'] == '2019-04-01')| 
                         (top_50_only['date'] == '2019-05-01')| 
                         (top_50_only['date'] == '2019-06-01')| 
                         (top_50_only['date'] == '2019-07-01')| 
                         (top_50_only['date'] == '2019-08-01')| 
                         (top_50_only['date'] == '2019-09-01')| 
                         (top_50_only['date'] == '2019-10-01')| 
                         (top_50_only['date'] == '2019-11-01')| 
                         (top_50_only['date'] == '2019-12-01')| 
                         (top_50_only['date'] == '2020-01-01')|
                         (top_50_only['date'] == '2020-02-01')|
                         (top_50_only['date'] == '2020-03-01')|
                         (top_50_only['date'] == '2020-04-01')|
                         (top_50_only['date'] == '2020-05-01')|
                         (top_50_only['date'] == '2020-06-01')|
                         (top_50_only['date'] == '2020-07-01')|
                         (top_50_only['date'] == '2020-08-01')|
                         (top_50_only['date'] == '2020-09-01')|
                         (top_50_only['date'] == '2020-10-01')|
                         (top_50_only['date'] == '2020-11-01')|
                         (top_50_only['date'] == '2020-12-01')|
                         (top_50_only['date'] == '2021-01-01')|
                         (top_50_only['date'] == '2021-02-01')|
                         (top_50_only['date'] == '2021-03-01')|
                         (top_50_only['date'] == '2021-04-01')|
                         (top_50_only['date'] == '2021-05-01')|
                         (top_50_only['date'] == '2021-06-01')|
                         (top_50_only['date'] == '2021-07-01')|
                         (top_50_only['date'] == '2021-08-01')|
                         (top_50_only['date'] == '2021-09-01')|
                         (top_50_only['date'] == '2021-10-01')]
                         
                         
```
